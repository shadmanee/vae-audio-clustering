```python
import random, numpy as np
from utils.audio_data import extract_mel_spectrogram
from config import config
from pathlib import Path
```

```python
root = Path("../..")
```

```python
EN_SOURCE_DIR = root / config.RAW_AUDIO_DIR_EN
BN_SOURCE_DIR = root / config.RAW_AUDIO_DIR_BN
FEATURES_DIR = root / config.FEATURES_DIR
```

```python
EN_SOURCE_DIR.exists(), BN_SOURCE_DIR.exists() # have to be true
```

```python
FEATURES_DIR.mkdir(exist_ok=True, parents=True)
```

```python
en_audio_list = [p for p in EN_SOURCE_DIR.iterdir() if p.is_file() and p.suffix.lower() == ".wav"]
if BN_SOURCE_DIR.exists():
    bn_audio_list = [p for p in BN_SOURCE_DIR.iterdir() if p.is_file() and p.suffix.lower() == ".wav"]
```

```python
audio_list = en_audio_list + bn_audio_list if "en_bn" in str(FEATURES_DIR) else en_audio_list
```

```python
FEATURES_DIR
```

```python
for audio_path in audio_list:
    try:
        spec = extract_mel_spectrogram(audio_path)
        out_path = FEATURES_DIR / f"{audio_path.stem}.npy"
        np.save(out_path, spec)
    except Exception as e:
        print(f"Error processing {audio_path.name}: {e}")
```

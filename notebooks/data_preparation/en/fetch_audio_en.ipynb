```python
from pathlib import Path
import shutil
import pandas as pd
```

```python
SOURCE_DIR = Path(r"D:\MSc\3. Spring 2026\CSE715\Project\MERGE_Audio_Balanced") # -> DO NOT REMOVE
# SOURCE_DIR = Path(input("Enter the path to MERGE_Audio_Balanced in your device: "))
DEST_DIR = Path("../../../data/audio/en/")
METADATA_FILE = Path("../../../data/metadata/metadata_en.csv")
```

```python
def find_audio_file(quadrant, stem):
    """
    Find the MP3 files in "<SOURCE_DIR>/<quadrant>/<stem>.mp3".
    """
    candidate = SOURCE_DIR / quadrant / f"{stem}.mp3"
    return candidate if candidate.exists() else None
```

```python
def transfer_audio_file(src, dst=DEST_DIR, copy_instead_of_move=True):
    """
    Creates destination folder if not already present. Copies or moves the MP3 file from the source folder to the destination folder.
    """
    dst.mkdir(parents=True, exist_ok=True)
    shutil.copy2(src, dst) if copy_instead_of_move else shutil.move(str(src), str(dst))
```

```python
metadata_df = pd.read_csv(METADATA_FILE)
metadata_df.head()
```

```python
found = 0
missing = []
```

```python
for _, row in metadata_df.iterrows():
    stem = str(row["audio_file_stem"]).strip()
    quadrant = str(row["quadrant"]).strip()
    try:
        src_file = find_audio_file(quadrant=quadrant, stem=stem)
        if src_file is None: 
            missing.append(stem)
            continue
        # transfer_audio_file(src=src_file)
        found += 1
    except Exception as e:
        print(f"{e}: {stem}")
```

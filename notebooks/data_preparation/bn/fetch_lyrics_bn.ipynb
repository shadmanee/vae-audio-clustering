```python
from pathlib import Path
import numpy as np
import pandas as pd
import shutil
```

```python
root = Path("../../..")
```

```python
DEST_DIR = root / config.LYRICS_DIR_BN
METADATA_FILE = root / config.METADATA_DIR / "bn/metadata_bn.csv"
FEATURES_DIR = root / config.FEATURES_DIR
```

```python
METADATA_FILE.exists(), FEATURES_DIR.exists() # have to be True
```

```python
DEST_DIR.mkdir(parents=True, exist_ok=True)
```

```python
all_feature_files = list(FEATURES_DIR.glob("*.npy"))
parent_ids = ["_".join(f.stem.split('_')[:-2]) for f in all_feature_files if len(f.stem.split("_")) > 2]
unique_song_ids = sorted(list(set(parent_ids)))

print(f"Total .npy files found: {len(all_feature_files)}")
print(f"Total unique songs extracted: {len(unique_song_ids)}")
print("First 10 unique IDs:", unique_song_ids[:10])
```

```python
df_meta = pd.read_csv(METADATA_FILE)
relevant_meta = df_meta[df_meta['id'].isin(unique_song_ids)]
```

```python
def save_to_txt(row):
    path = f"{DEST_DIR}/{row['id']}.txt"
    with open(path, "w", encoding="utf-8") as f:
        f.write(str(row['sentence']))
```

```python
relevant_meta.apply(save_to_txt, axis=1)
```

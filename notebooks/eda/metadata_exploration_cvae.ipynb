```python
import pandas as pd
from pathlib import Path
```

```python
from config import config
```

```python
root = Path("../..")
```

```python
en_meta_path = root / config.METADATA_DIR / "metadata_en.csv"
bn_meta_path = root / config.METADATA_DIR / "bn/metadata_bn.csv"
en_meta_path.exists(), bn_meta_path.exists() # has to be True
```

```python
en_meta_df = pd.read_csv(en_meta_path)
bn_meta_df = pd.read_csv(bn_meta_path)
```

```python
en_meta_df.head()
```

```python
bn_meta_df.head()
```

```python
en_meta_df["language"] = "en"
en_meta_df["label"] = 0
bn_meta_df["language"] = "bn"
bn_meta_df["label"] = 1
bn_meta_df.rename(columns={"id": "audio_file_stem"}, inplace=True)
```

```python
bn_meta_df
```

```python
combined_meta_df = pd.concat([en_meta_df[["audio_file_stem", "language", "label"]], bn_meta_df[["audio_file_stem", "language", "label"]]], axis=0)
combined_meta_df.reset_index(drop=True, inplace=True)
```

```python
combined_meta_df
```

```python
DEST_PATH = root / config.METADATA_DIR / "metadata_en_bn.csv"
combined_meta_df.to_csv(DEST_PATH, index=False)
```
